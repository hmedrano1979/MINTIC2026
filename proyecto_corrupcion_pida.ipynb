Este notebook contiene el desarrollo técnico completo para la detección de anomalías y patrones de riesgo en el dataset PIDA.

## 1. Carga de Datos y Librerías
Conectamos directamente con la API de Datos Abiertos de Colombia.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import requests
from sklearn.ensemble import IsolationForest

url = "https://www.datos.gov.co/resource/k5ma-yb9b.json?$limit=5000"
df = pd.DataFrame(requests.get(url).json())
print(f"Dataset cargado: {df.shape[0]} registros")
df.head()
```

## 2. Limpieza Técnica
Tratamos nulos y preparamos las variables para el análisis.

```python
df = df.dropna(axis=1, thresh=int(0.5*len(df)))
for col in df.select_dtypes(include=['object']).columns:
    df[col] = df[col].fillna('N/A')
print("Limpieza completada.")
```

## 3. Modelo de IA: Detección de Anomalías
Usamos Isolation Forest para identificar comportamientos atípicos que puedan indicar riesgos de corrupción.

```python
# Codificación rápida para el modelo
df_encoded = df.copy()
for col in df_encoded.select_dtypes(include=['object']).columns:
    df_encoded[col] = df_encoded[col].astype('category').cat.codes

model = IsolationForest(contamination=0.05, random_state=42)
df['score_anomalia'] = model.fit_predict(df_encoded.select_dtypes(include=[np.number]))
print(f"Anomalías detectadas: {len(df[df['score_anomalia'] == -1])}")
```
